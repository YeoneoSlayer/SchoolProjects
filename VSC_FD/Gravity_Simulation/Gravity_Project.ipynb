```python
import pygame
import random
import math
```

```python

base_particles = []
updated_particles = []

pygame.init()

width = 1000
height = 800
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Gravity")
```

```python
# 입자 생성 함수
def particle_geration(): 
    x = random.randint(0,width)
    y = random.randint(0,height)
    vx = random.uniform(-0.3, 0.3)
    vy = random.uniform(-0.3, 0.3)
    m = random.uniform(0.5 ,5)
    c = (255,255,255)
    
    return x, y, vx, vy, m, c

# 가속도 계산 함수
def acceleration(x1, y1, m1, x2, y2, m2): 
    G = 0.002
    k = 0.001
    r = (((x1 - x2)**2 + (y1 - y2)**2)**0.5)*2.7
    
    if r == 0:
        return 0
    mo1 = (m1**2)*3.14
    mo2 = (m2**2)*3.14
    a = G * m1* m2 / (r**2)
    
    return a

# 가속도 분해 함수
def vecter_con(acceleration, x1, y1, x2, y2): 
    angle = math.atan2(y2 - y1, x2 - x1)  # 역탄젠트를 이용하여 두 입자 간 라디안 각도 추출
    xa = acceleration * math.cos(angle)  
    ya = acceleration * math.sin(angle) 
    return xa, ya

# 입자 가속도 적용 함수
def update_particle(particle): 
    
    x, y, vx, vy, m, c = particle
    
    # 속도에 가속도를 적용
    for other_particle in base_particles:
        if other_particle != particle:
            ox, oy, ovx, ovy, om, c = other_particle
            ax, ay = vecter_con(acceleration(x, y, m, ox, oy, om), x, y, ox, oy)
            vx += ax
            vy += ay
        else:
            pass
    
    # 위치 업데이트
    x += vx
    y += vy
    
    # 새로운 입자 정보 반환
    return x, y, vx, vy, m, c

# 입자 그리기 함수
def draw_particle(particle): 
    x, y, vx, vy, m, c = particle
    pygame.draw.circle(screen, c, (int(x), int(y)), m)
```

```python
# range 범위 만큼의 입자를 생성
for i in range(100): 
    particle = particle_geration()
    base_particles.append(particle)
```

```python

Value = True
# 프로그램 실행을 위한 반복문
while Value: 
    screen.fill((0, 0, 0))
    net_acc = 0

    
    for count in range(len(base_particles)):  # 입자 그리기
        draw_particle(base_particles[count])

    for world_count in range(len(base_particles)):  # 비교되는 입자 호출
        main_particle = base_particles[world_count]
        x1, y1, vx1, vy1, m1, c1 = main_particle
        for local_count in range(len(base_particles)):  # 비교할 입자 호출
            com_particle = base_particles[local_count]
            x2, y2, vx2, vy2, m2, c2 = com_particle

            # 합성 가속도 구하기
            acc = acceleration(x1, y1, m1, x2, y2, m2)
            net_acc += acc

            # 가속도 분해
            xa, ya = vecter_con(net_acc, x1, y1, x2, y2)

        vx1 += xa*20
        vy1 += ya*20

        x1 += vx1
        y1 += vy1
        
        updated_particle = x1, y1, vx1, vy1, m1, c1
        updated_particles.append(updated_particle)

    base_particles = updated_particles
    updated_particles = []  # updated_particles 리스트 초기화

    for event in pygame.event.get(): #탈출 구문
        if event.type == pygame.QUIT:
            Value = False

    pygame.display.update() # 화면 업데이트

pygame.quit()
```
