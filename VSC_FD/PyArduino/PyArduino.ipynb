```python
#import serial
import time
import cvlib as cv
from cvlib.object_detection import draw_bbox
import cv2


# Serial 함수를 이용해서 port와 baudrate를 설정

'''
py_serial = serial.Serial(
    
    # Window
    port='COM13',
    
    # 보드 레이트 (통신 속도)
    baudrate=9600,
)

'''

# 웹캠 열기 1
webcam = cv2.VideoCapture(0)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()
    

# loop through frames 1
while webcam.isOpened():

    # read frame from webcam  1
    status, frame = webcam.read()

    if not status:
        break
        
        
    # 물체 검출
    bbox, label, conf = cv.detect_common_objects(frame)

    print(bbox, label, conf)

    # 검출된 물체 가장자리에 바운딩 박스 그리기
    out = draw_bbox(frame, bbox, label, conf, write_conf=True)

    # 화면 출력
    cv2.imshow("CCTV", out)
    
    # 카운트 된 사람의 수
    num = label.count('person')
    
    
    # 카운트 된 사람에 따른 조건문
    if num >= 15:
        print("15 이상 감지되었습니다")
        
        
        # 아두이노와의 통신
        commend = input('입력 필요')
    
        py_serial.write(commend.encode())
        
        # 0.1 초간 대기
        time.sleep(0.1)
        
        

    # S 키를 누르면 정지
    if cv2.waitKey(1) & 0xFF == ord('s'):
        break

    
# release resources
webcam.release()
cv2.destroyAllWindows()
```
